# Slide 11
<li>Plotting histograms, only for numeric attributes.</li>

```python
import matplotlib.pyplot as plt
```

```python
def mkdir_p(mypath):

    from errno import EEXIST
    from os import makedirs,path

    try:
        makedirs(mypath)
    except OSError as exc: 
        if exc.errno == EEXIST and path.isdir(mypath):
            pass
        else: raise
```

```python
binn = 8
for (name, series) in df.iteritems():
    if series.dtype != 'object':
        ax = df[name].plot.hist(bins=binn)
        ax.set_title('Attribute: '+df[name].name+ '- Histogram with '+str(binn)+ ' Bins')
        ax.set_xlabel(df[name].name)
        ax.set_ylabel('Count')
        fig = ax.figure
        fig.set_size_inches(8,3)
        fig.tight_layout(pad=1)
        temp_path = "Figures//Slide 11//Figures with Bins "+str(binn)
        mkdir_p(temp_path)
        fig.savefig(temp_path + "//"+ df[name].name+'.png',dpi=600)
        plt.close(fig)
        print("Plotting for attribute "+df[name].name + " with " + str(binn) +" bins has been completed\n")
```

# Slide 20
<li>Min-Max normalization to [0-1].</li>

```python
from sklearn import preprocessing
numericDF=df._get_numeric_data()
print("Before Normalization:")
quantitative_DF=numericDF.drop(['ID','Fraud Flag','Num Soft Tissue'], axis=1)
print(quantitative_DF.head(10))
print(quantitative_DF.shape)

print("\nAfter Normalization:")
x=quantitative_DF.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled=min_max_scaler.fit_transform(x)
quantitative_DF=pd.DataFrame(x_scaled,columns=quantitative_DF.columns)
print(quantitative_DF.head(10))
print(quantitative_DF.shape)
```

# Slide 21
<li>Z-score normalization.</li>

```python
from scipy.stats import zscore

print("Pre-Normalization:")
quantitative_DF=numericDF.drop(['ID','Fraud Flag','Num Soft Tissue'], axis=1)
print(quantitative_DF.head(10))
print(quantitative_DF.shape)

print("Post-Normalization:")
quantitative_DF=quantitative_DF.apply(zscore)
print(quantitative_DF.head(10))
print(quantitative_DF.shape)
```

# Slide 22-24
<li>Equal-Width Binning.</li>

```python
print("Data for Binning:")
quantitative_DF=numericDF.drop(['ID','Fraud Flag','Num Soft Tissue'], axis=1)
print(quantitative_DF.head(10))
print(quantitative_DF.shape)

print("After Equal-Width Binning:")
for (name,series) in quantitative_DF.iteritems():
    count, division = np.histogram(series, bins=np.linspace(series.min(),series.max(),5))
    print("\n")
    print("Attribute:", name)
    print('Border of Bins:', division)
    print('Frequency of respective bins:', count)
```
